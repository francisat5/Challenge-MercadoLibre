```
import requests
import json
```

```
url = 'https://api.mercadolibre.com/sites/MLA/search?q=chromecast&limit=50#json'
response = requests.get(url)
data=response.json()
aux=data["results"]
```

```
listaId = []
for item in aux:
    listaId.append(list(item.values())[0])
print(listaId)
```

```
datos = []
for id in listaId:
    url = 'https://api.mercadolibre.com/items/' + id
    response = requests.get(url)
    datos.append(response.json())
```

```
print (len(datos))
print (datos[0])
```
